```python
#importing pandas and numpy libraries
import pandas as pd

import numpy as np
```

```python
#loading our dataset
df = pd.read_csv('http://bit.ly/Autolibdataset')
df.head()
```

```python
#previewing the first 10 rows in the dataset
df.head(10)
```

```python
#discovering information about dataset
df.info()
```

```python
#dropping the columns that are irrelevant
df = df.drop(['Cars', 'Geo point', 'Address', 'Displayed comment', 'Charging Status', 'Subscription status', 'year', 'month', 'ID'], axis=1)
df.head()
```

```python
df.drop(['Scheduled at'], axis=1, inplace=True)
df.info()
```

```python
df.drop(['Charge Slots'], axis=1, inplace=True)

df.info()
```

```python
#checking the completeness of the dataset and improving it
df.isnull() .sum()
```

```python
df.notnull().sum()
```

```python
df.head()
```

```python
#improving the consistency by checking for duplicates
df.duplicated().sum()
```

```python
#dropping the duplicated values
df.drop_duplicates(inplace=True)
```

```python
#confirming if the duplicated calues have been dropped
df.duplicated().sum()
```

```python
#checking the number of rows and columns in the dataset
df.shape
```

```python
#checking the column names
df.columns
```

```python
df.head()
```

```python
#Most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018
crrr = df[df['City'] == 'Paris']
rac = crrr.groupby(['Public name', 'hour'])[['Bluecar counter']].sum()
rac.head()
```

```python
rac['Bluecar_usage'] = rac['Bluecar counter'].diff()

rac.head()
```

```python
rc = rac[rac['Bluecar_usage'] < 0]
rc.head()
```

```python
rccc = rc.groupby('hour')[['Bluecar_usage']].sum()
rccc
```

```python
rccc.sort_values('Bluecar_usage', ascending=True)
```

```python
#creating a new column to hold the difference in cars at the station
#finding the difference in the number of cars at the station
df['Bluecar_usage'] = df['Bluecar counter'].diff()
```

```python
# creating column for utilib usage
df['Utilib_usage'] = df['Utilib counter'].diff()

# creating column for utilib 14 usage
df['Utilib_14_usage'] = df['Utilib 1.4 counter'].diff()

df.head()
```

```python
#most popular hour for picking up a blue car
df = df[(df['City'] == 'Paris') & (df['Bluecar_usage'] < 0)]

df.head()


car = df.groupby(['hour'])[['Bluecar_usage']].sum()
car.head()
```

```python
carr = car.sort_values('Bluecar_usage', ascending=True)
carr.head()
```
